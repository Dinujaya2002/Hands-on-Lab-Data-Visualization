# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the 'ConvertedComp' column from the 'master' table
QUERY = """
SELECT ConvertedComp
FROM master
WHERE ConvertedComp IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot a histogram of the ConvertedComp column
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the 'Age' column from the 'master' table
QUERY = """
SELECT Age
FROM master
WHERE Age IS NOT NULL AND Age > 0
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot a box plot of the Age column
plt.figure(figsize=(8, 6))
plt.boxplot(df['Age'], vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the 'Age' and 'WorkWeekHrs' columns from the 'master' table
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
WHERE Age IS NOT NULL AND WorkWeekHrs IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot a scatter plot of Age vs. WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5, color='blue')
plt.title('Scatter Plot of Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' columns from the 'master' table
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE WorkWeekHrs IS NOT NULL AND CodeRevHrs IS NOT NULL AND Age IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot a bubble plot of WorkWeekHrs vs. CodeRevHrs with Age as bubble size
plt.figure(figsize=(12, 8))
bubble_size = df['Age'] / df['Age'].max() * 100  # Normalize Age for bubble size
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_size, alpha=0.5, color='green')

plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs (Bubble Size: Age)')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the table schema
schema_query = """
PRAGMA table_info(master);
"""
df_schema = pd.read_sql_query(schema_query, conn)

# Display the schema
print("Table Schema:")
print(df_schema)

# Identify the correct column name for database preferences
# Replace 'DatabaseLearnNextYear' with the actual column name found in the schema
column_name = 'DatabaseLearnNextYear'  # Update this based on schema results

# Query to get the database preferences
data_query = f"""
SELECT {column_name}
FROM master
WHERE {column_name} IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(data_query, conn)

# Close the database connection
conn.close()

# Get the top 5 databases
top_databases = df[column_name].value_counts().head(5)

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    top_databases,
    labels=top_databases.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.get_cmap('tab20').colors
)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the WorkWeekHrs and CodeRevHrs for the age group 30 to 35
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
AND WorkWeekHrs IS NOT NULL
AND CodeRevHrs IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Calculate median values
median_workweek_hrs = df['WorkWeekHrs'].median()
median_coderev_hrs = df['CodeRevHrs'].median()

# Prepare data for stacked bar chart
data = pd.DataFrame({
    'Category': ['WorkWeekHrs', 'CodeRevHrs'],
    'Median Hours': [median_workweek_hrs, median_coderev_hrs]
})

# Plot a stacked bar chart
plt.figure(figsize=(8, 6))
plt.bar(data['Category'], data['Median Hours'], color=['skyblue', 'salmon'])
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30 to 35')
plt.xlabel('Category')
plt.ylabel('Median Hours')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the ConvertedComp for ages between 45 and 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
AND ConvertedComp IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Calculate median ConvertedComp for each age
median_compensation = df.groupby('Age')['ConvertedComp'].median().reset_index()

# Plot the median ConvertedComp for each age
plt.figure(figsize=(10, 6))
plt.plot(median_compensation['Age'], median_compensation['ConvertedComp'], marker='o', linestyle='-', color='b')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.grid(True)
plt.xticks(range(45, 61))  # Ensures all ages from 45 to 60 are shown on x-axis
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the MainBranch data
QUERY = """
SELECT MainBranch
FROM master
WHERE MainBranch IS NOT NULL
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Count the frequency of each value in MainBranch
mainbranch_counts = df['MainBranch'].value_counts()

# Plot a horizontal bar chart
plt.figure(figsize=(10, 6))
mainbranch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
